```python
def read_or_load_excel(file_path,save_path,sheet_name=None):
  try:
    df=pd.read_pickle(save_path)
  except FileNotFoundError:
    df= pd.read_excel(file_path,dtype=str,sheet_name=sheet_name)
    df.to_pickle(save_path)
  return df
```
